```python
import numpy as np, matplotlib.pyplot as plt, matplotlib as mpl
import pcmdpy as ppy
%matplotlib inline
```

```python
a = ppy.sfhmodels.ConstantSFR(np.array([3.25])).as_default()
gal_nonparam = ppy.galaxy.CustomGalaxy(
    ppy.metalmodels.FixedWidthNormMDF(0.5),
    ppy.dustmodels.FixedWidthLogNormDust(0.15),
    ppy.sfhmodels.NonParam(),
    ppy.distancemodels.VariableDistance()
)
gal_nonparam.set_params([-0.2, -0.5] + list(np.log10(a.as_default().SFH)) + [29.2])
```

```python
res = ppy.results.ResultsPlotter('results/mock_SFH5.csv', true_model=gal_nonparam)
```
