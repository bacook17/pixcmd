```python
import numpy as np, matplotlib.pyplot as plt, matplotlib as mpl
import pcmdpy as ppy
from astropy.io import fits
%matplotlib inline
```

```python
res = {}
```

```python
regions = []
for r in range(25):
    try:
        res['r{:d}'.format(r)] = ppy.plotting.ResultsPlotter('results/m49_taufull_r{:d}.csv'.format(r),
                                                            run_name='Region {:d}'.format(r))
        res['r{:d}_nm'.format(r)] = ppy.plotting.ResultsPlotter('results/m49_taunomu_r{:d}.csv'.format(r),
                                                            run_name='Region {:d} (No Mu)'.format(r))
        res['r{:d}_wd'.format(r)] = ppy.plotting.ResultsPlotter('results/m49_tauwide_r{:d}.csv'.format(r),
                                                            run_name='Region {:d} (Wide)'.format(r))
        regions.append(r)
    except:
        print('r{:d} not found'.format(r))
```

```python
xmax = np.max([len(r.df) for r in res.values()])
print(xmax)
```

```python
plt.plot(res['r3'].df['logfeh_mean'].iloc[2000:2050])
```

```python
fig, axes = plt.subplots(ncols=3*len(regions), nrows=8, figsize=(30,15), sharex=True, sharey='row')
for i, r in enumerate(regions):
    res['r{:d}'.format(r)].plot_chains(axes=axes[:,3*i])
    res['r{:d}_wd'.format(r)].plot_chains(axes=axes[:,3*i+1])
    res['r{:d}_nm'.format(r)].plot_chains(axes=axes[np.array([0,1,2,4,5,6,7]),3*i+2])
    for j in range(3):
        axes[3,j+3*i].axhline(y=31.3, ls='--', color='r')
axes[0,0].set_xlim(xmax=xmax)
plt.tight_layout()
```
