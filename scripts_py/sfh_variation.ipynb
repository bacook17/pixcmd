```python
import numpy as np, matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
d_param = np.array([-.5, -.25, -.1, 0., .1, .25, .5])
```

```python
chi2 = {}
chi2['feh'] = np.array([8.3e5, 9.3e5, 4.8e5, 7.0e2, 1.8e5, 6.6e5, 1.0e6])
chi2['dust'] = np.array([6.9e5, 4.6e5, 1.6e5, 7.0e2, 2.2e5, 7.5e5, 1.0e6])
chi2['sfh0'] = np.array([6.5e4, 2.6e4, 6.1e3, 7.0e2, 9.9e3, 5.9e4, 2.9e5])
chi2['sfh1'] = np.array([3.7e5, 1.7e5, 4.2e4, 7.0e2, 6.0e4, 3.2e5, 7.8e5])
chi2['sfh2'] = np.array([2.2e5, 1.1e5, 2.4e4, 7.0e2, 4.1e4, 2.6e5, 7.4e5])
chi2['sfh3'] = np.array([5.7e5, 3.3e5, 9.9e4, 7.0e2, 1.5e5, 6.3e5, 9.8e5])
chi2['sfh4'] = np.array([5.7e5, 3.1e5, 8.5e4, 7.0e2, 1.3e5, 6.1e5, 1.0e6])
chi2['sfh5'] = np.array([9.0e5, 6.9e5, 2.8e5, 7.0e2, 3.7e5, 9.2e5, 1.0e6])
chi2['sfh6'] = np.array([5.0e5, 2.7e5, 7.9e4, 7.0e2, 1.2e5, 5.5e5, 9.7e5])
```

```python
for k, v in chi2.items():
    plt.plot(d_param, v, label=k, ls='-', marker='o')
plt.xlabel(r'$\Delta$ X (dex)')
plt.ylabel(r'$\chi^2$')
plt.yscale('log')
plt.legend(loc=0)
```
