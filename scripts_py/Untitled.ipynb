```python
import numpy as np, matplotlib.pyplot as plt, matplotlib as mpl
%matplotlib inline
```

```python
plt.style.use('seaborn-poster')
```

```python
def func_0(m, d, nm, nd):
    err = np.sqrt(2 * (m+d))
    err += 2.*np.exp(-err)
    return (m*nd - d*nm) / err
```

```python
def func_1(m, d, nm, nd):
    return np.log(nd*nm + nd*m) - np.log(nd*nm + nm*d)
```

```python
def func_2(m,d,nm,nd):
    return np.log(np.sqrt(nd*nm) + nd*m) - np.log(np.sqrt(nd*nm) + nm*d)
```

```python
def func_3(m,d,nm,nd):
    return np.log(1 + nd*m) - np.log(1 + nm*d)
```

```python
def func_4(m,d,nm,nd):
    return np.log(nd + nd*m) - np.log(nm + nm*d)
```

```python
str_names = {}
str_names[func_0] = r'$\chi$'
str_names[func_1] = r'$\log\left(\frac{n_d n_m + n_d m}{n_d n_m + n_m d}\right)$'
str_names[func_2] = r'$\log\left(\frac{\sqrt{n_d n_m} + n_d m}{\sqrt{n_d n_m} + n_m d}\right)$'
str_names[func_3] = r'$\log\left(\frac{1 + n_d m}{1 + n_m d}\right)$'
str_names[func_4] = r'$\log\left(\frac{n_d + n_d m}{n_m + n_m d}\right)$'
```

```python
def make_plot(nm, nd, func, c_max=100):
    fig, ax = plt.subplots(figsize=(6,5))
    ms = np.arange(int(c_max * nm / nd))
    ds = np.arange(c_max)
    x = np.array([[func(m, d, nm, nd) for d in ds] for m in ms])
    x_max = np.max(np.abs(x))
    plt.subplot(ax)
    plt.imshow(x, cmap='bwr_r', interpolation='none',
               norm=mpl.colors.Normalize(vmin=-x_max, vmax=x_max), origin='lower',
              aspect='auto')
    ax.set_xlabel('data counts')
    ax.set_ylabel('model counts')
    plt.tight_layout()
    ax.set_title(str_names[func])
    plt.colorbar()
    return ax
```

```python
make_plot(128*128, 128*128, func_0)
```

```python
make_plot(512*512, 128*128, func_0)
```

```python
make_plot(128*128, 128*128, func_1)
```

```python
make_plot(512*512, 128*128, func_1)
```

```python
make_plot(128*128, 128*128, func_2)
```

```python
make_plot(512*512, 128*128, func_2)
```

```python
make_plot(128*128, 128*128, func_3)
```

```python
make_plot(512*512, 128*128, func_3)
```

```python
make_plot(128*128, 128*128, func_4, c_max=10)
```

```python
make_plot(512*512, 128*128, func_4, c_max=10)
```
