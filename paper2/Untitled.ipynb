```python
import numpy as np, matplotlib as mpl, matplotlib.pyplot as plt
import pcmdpy as ppy
%matplotlib inline
```

```python
m1 = ppy.metalmodels.FixedWidthNormMDF(sig=0.5, initial_params=np.array([0.]))
m2 = ppy.metalmodels.ClosedBoxMDF(initial_params=np.array([0.]))
```

```python
plt.plot(m1.fehs, m1.weights, 'k-')
```

```python
plt.plot(m1.fehs, m1.weights, 'k-')
plt.plot(m2.fehs, m2.weights, 'r-')
```

```python
len(m2.fehs), len(m1.fehs)
```

```python
s1 = ppy.sfhmodels.TauModel(np.array([0.0, 5.0]), iso_step=-1)
s2 = ppy.sfhmodels.RisingTau(np.array([0.0, 5.0]), iso_step=-1)
```

```python
for t in [0.5, 1.0, 2.0, 3.0, 5.0, 10.0]:
    s1.set_params(np.array([0., t]))
    s2.set_params(np.array([0., t]))    
    plt.plot(s1.ages, s1.logSFH, ls='--')
    plt.plot(s2.ages, s2.logSFH, ls='-')
```
