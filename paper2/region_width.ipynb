```python
import numpy as np, matplotlib.pyplot as plt
import pcmdpy as ppy
%matplotlib inline
```

```python
pcmd1 = np.loadtxt('../data/M87v2/pcmds/M87_I_gI_128.pcmd', unpack=True)
pcmd2 = np.loadtxt('../data/M87v2/pcmds/M87_I_gI_wide.pcmd', unpack=True)
pcmd3 = np.loadtxt('../data/M87v2/pcmds/M87_I_gI_quad.pcmd', unpack=True)
```

```python
pcmd4 = np.loadtxt('../data/M87v2/pcmds/M87_I_gI_125.pcmd', unpack=True)
pcmd5 = np.loadtxt('../data/M87v2/pcmds/M87_I_gI_126.pcmd', unpack=True)
pcmd6 = np.loadtxt('../data/M87v2/pcmds/M87_I_gI_127.pcmd', unpack=True)
```

```python
f = ppy.instrument.m87_filters_v2()
iso_model = ppy.isochrones.Isochrone_Model(f)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ppy.plotting.plot_pcmd_contours(pcmd1, ax=ax, color='k', label='Default', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_pcmd_contours(pcmd2, ax=ax, color='r', label='Wide Slice', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_pcmd_contours(pcmd3, ax=ax, color='b', label='Full Ellipse', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_isochrone(iso_model, axes=ax, alpha=0.)
ax.set_xlim([0.5, 2.5])
ax.legend(loc=0)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ppy.plotting.plot_pcmd_contours(pcmd1, ax=ax, color='k', label='Q1', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_pcmd_contours(pcmd4, ax=ax, color='r', label='Q2', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_pcmd_contours(pcmd5, ax=ax, color='b', label='Q3', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_pcmd_contours(pcmd6, ax=ax, color='g', label='Q4', span=[[0.5, 2.5],[25.5, 26.5]])
ppy.plotting.plot_isochrone(iso_model, axes=ax, alpha=0.)
ax.set_xlim([0.5, 2.5])
ax.legend(loc=0)
fig.tight_layout()
```
